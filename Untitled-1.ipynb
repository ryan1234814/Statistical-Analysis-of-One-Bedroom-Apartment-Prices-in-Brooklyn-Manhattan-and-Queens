```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Example data
brooklyn_price = np.array([3200, 3300, 3100, 2900, 2800, 3400, 3500, 3300])
manhattan_price = np.array([4300, 4200, 4000, 3900, 4100, 4300, 4400, 4200])
queens_price = np.array([2800, 2700, 2600, 2400, 2500, 2600, 2700, 2500])
data={
    'Brooklyn':brooklyn_price,
    'Manhattan':manhattan_price,
    'Queens':queens_price
}
df=pd.DataFrame(data)
print(df)
brooklyn_mean=np.mean(brooklyn_price)
manhattan_mean=np.mean(manhattan_price)
queens_mean=np.mean(queens_price)

brooklyn_median=np.median(brooklyn_price)
manhattan_median=np.median(manhattan_price)
queens_median=np.median(queens_price)
brooklyn_mode = stats.mode(brooklyn_price, keepdims=True).mode[0]
manhattan_mode = stats.mode(manhattan_price, keepdims=True).mode[0]
queens_mode = stats.mode(queens_price, keepdims=True).mode[0]


print("Brooklyn - Mean:", brooklyn_mean, "Median:", brooklyn_median, "Mode:", brooklyn_mode)
print("Manhattan - Mean:", manhattan_mean, "Median:", manhattan_median, "Mode:", manhattan_mode)
print("Queens - Mean:", queens_mean, "Median:", queens_median, "Mode:", queens_mode)

df.hist(bins=5,figsize=(10,8))
plt.suptitle('Histogram of One bedroom apartment prices')
plt.show()
df.plot(kind='bar',figsize=(10,6))
plt.suptitle('Apartment Prices by Borough')
plt.xlabel('Index')
plt.ylabel('Price ($)')
plt.show()
mean_values=[brooklyn_mean,manhattan_mean,queens_mean]
boroughs=['Brooklyn','Manhattan','Queen']
plt.pie(mean_values,labels=boroughs,startangle=140)
plt.title('Mean Price distribution Prices by Borough')
plt.show()
t_test_brooklyn_manhattan=stats.ttest_ind(brooklyn_price,manhattan_price)
t_test_brooklyn_queens=stats.ttest_ind(brooklyn_price,queens_price)
t_test_manhattan_queens=stats.ttest_ind(manhattan_price,queens_price)
print("Brooklyn vs Manhattan p-value:", t_test_brooklyn_manhattan.pvalue)
print("Brooklyn vs Queens p-value:", t_test_brooklyn_queens.pvalue)
print("Manhattan vs Queens p-value:", t_test_manhattan_queens.pvalue)
```
